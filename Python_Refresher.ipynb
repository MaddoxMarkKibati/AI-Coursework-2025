### Program to print out the number of days

```python
dayseconds = int(86400) # Variable containing the number of seconds in a day
noofdays = int(input("Hello, how many days would like to know the numbers of seconds for?: ")) # Variable and prompt for the number of days
print("There are ", dayseconds * noofdays, "seconds in ", noofdays, "days.") # Output for the number of seconds in the day amount recieved from the user
```

### Program to calculate the volume of a sphere

```python
import math # Import needed for the pi
radius = int(input("Please input the radius of the sphere: ")) # Prompt for the radius and the variable to store it
volume = int((4/3) * math.pi * (radius ** 3)) # Code to calculate the volume and store it in a variable 
print("The volume of a sphere of radius ", radius, " is ", volume) # Code to print out the volume to the user (wasnt required by the question)
```

### Program to compute the area of a square

```python
length = int(input("Hello, what is the length of the square?: ")) # Prompt for the side length of the square and the vriable to store it
perimeter = int(length * 4) # Code to calculate the perimter and store it as a variable
area = int(length ** 2) # Code to calculate the area and store it as a variable
print("Perimeter =", perimeter, "\nArea =", area) # Code to output of the Area and Perimeter of the square
```

### Program to determine whether character input is usercase or lowercase

```python
def checkcase(char):
  if 'a' <= char <= 'z':
    return "lowercase"
  elif 'A' <= char <= 'Z':
    return "uppercase"
  else:
    return "not a letter"
character = input("Enter a single character: ")
result = checkcase(character)

if result == "lowercase":
  print("The character ", character, " is a lowercase letter.")
elif result == "uppercase":
  print("The character ", character, " is an uppercase letter.")
else:
  print("The character ", character, " is not a letter.")
```

# Python Program of the Pseudocode

```python
# Initialisation of Variable
x = 0
y = 20

while y >= 6:
    y -= 4
    if y != 0:
        x += 2 / y

print(x)
```
